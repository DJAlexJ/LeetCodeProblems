```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
    def append(self, element):
        current = self
        if current:
            while current.next:
                current = current.next
            current.next = ListNode(element)
        else:
            self.val = element
            self.next = None
        
    def print_(self):
        current = self
        if current:
            while current.next:
                print(current.val)
                current = current.next
            print(current.val)
        else:
            print('-')
```

```python
#42. Trapping Rain Water (Hard)
class Solution:
    def trap(self, height: List[int]) -> int:
        stack = []
        drops = 0
        current = 0
        while current < len(height):
            while stack != [] and height[current] > height[stack[-1]]:
                top = stack[-1]
                stack.remove(stack[-1])
                if stack == []:
                    break
                distance = current - stack[-1] - 1
                bounded_height = min(height[current], height[stack[-1]]) - height[top]
                drops += distance * bounded_height
            stack.append(current)
            current += 1
        return drops
```

```python
#23. Merge k Sorted Lists (Hard)
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        res = ListNode(0)
        aux = res
        values = []
        for lst in lists:
            while lst:
                values.append(lst.val)
                lst = lst.next
        values = sorted(values)
        for val in values:
            aux.next = ListNode(val)
            aux = aux.next
        return res.next
```

```python
#41. First Missing Positive (Hard)
class Solution:
    def firstMissingPositive(self, nums: List[int]) -> int:
        hash_map={}
        for i in range(len(nums)):
            if nums[i]>0:
                hash_map[nums[i]]=i       
        j=1
        while True:
            if j not in hash_map:
                return j 
            j+=1
```

```python
#99. Recover Binary Search Tree (Hard)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        res, stack, first, second = None, [], None, None
        while True:
            while root:
                stack.append(root)
                root = root.left
            if stack == []:
                break 
            node = stack.pop()
        # first time occurs reversed order
            if res and res.val > node.val:
                if first is None:
                     first = res
            # first or second time occurs reversed order
                second = node
            res = node
            root = node.right
        first.val, second.val = second.val, first.val
```

```python
#295. Find Median from Data Stream (Hard)
class MedianFinder:
    def __init__(self):
        self.arr = []
    
    def addNum(self, num: int) -> None:
        left, right = 0, len(self.arr)
        while left < right:
            mid = left + (right - left) // 2
            if self.arr[mid] < num:
                left = mid + 1
            else:
                right = mid
        self.arr.insert(left, num)
        
    def findMedian(self) -> float:
        n = len(self.arr)
        if n % 2 == 0:
            return (self.arr[n//2-1] + self.arr[n//2]) * 0.5
        else:
            return self.arr[n//2]
```

```python
#312. Burst Balloons (Hard)
class Solution:
    def maxCoins(self, nums: List[int]) -> int:
        nums = [1] + nums + [1]
        dp = [[0]*len(nums) for i in range(len(nums))]
        
        for gap in range(0,len(nums)-2):
            for i in range(1,len(nums)-1-gap):
                j = i+gap
                for k in range(i,j+1):
                    a = dp[i][k-1] if k>i else 0
                    b = dp[k+1][j] if k<j else 0
                    dp[i][j] = max(dp[i][j],nums[i-1]*nums[k]*nums[j+1] + a + b)
                    
        return dp[1][len(nums)-2]
```

```python
#329. Longest Increasing Path in a Matrix (Hard)

#My Solution (TimeLimitExceeded?)
class Solution:
    def longestPath(self, i, j):
        if i < 0 or i >= len(self.matrix) or j < 0 or j >= len(self.matrix[i]):
            return -1
        else:
            val = self.matrix[i][j]
            return max(self.longestPath(i-1, j) if i >= 1 and  val < self.matrix[i-1][j] else 0,
                       self.longestPath(i+1, j) if i <= len(self.matrix)-2 and val < self.matrix[i+1][j] else 0,
                       self.longestPath(i, j+1) if j <= len(self.matrix[i])-2 and val < self.matrix[i][j+1] else 0,
                       self.longestPath(i, j-1) if j >= 1 and val < self.matrix[i][j-1] else 0) + 1
    
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        self.matrix = matrix
        if len(matrix) == 0 or len(matrix[0]) == 0:
            return 0
        res = 0
        for i in range(len(matrix)):
            for j in range(len(matrix[i])):
                res = max(self.longestPath(i, j), res)
        return res

#Solution
class Solution:
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        res = 0
        self.matrix = matrix
        self.memo = [[0] * len(matrix[0]) for i in range(len(matrix))]
        
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                res = max(self.dfs(i, j), res)
        
        return res
    
    def dfs(self, i: int, j: int) -> int:
        if not self.memo[i][j]:
            val = self.matrix[i][j]
            self.memo[i][j] = 1 + \
                            max(self.dfs(i-1, j) if i and val < self.matrix[i-1][j] else 0,
                                self.dfs(i+1, j) if i < len(self.matrix)-1 and val < self.matrix[i+1][j] else 0,
                                self.dfs(i, j-1) if j and val < self.matrix[i][j-1] else 0,
                                self.dfs(i, j+1) if j < len(self.matrix[0])-1 and val < self.matrix[i][j+1] else 0)
        return self.memo[i][j]
```

```python
#493. Reverse Pairs (Hard)
from bisect import bisect, insort

class Solution(object):
    def reversePairs(self, nums):
        lens, running_nums = [0]*len(nums), [] 
        for i in range(len(nums)):
            lens[i] = len(running_nums) - bisect(running_nums, 2*nums[i])
            insort(running_nums, nums[i])
        
        return sum(lens)
```

```python
#739. Daily Temperatures (Medium)
class Solution:
    def dailyTemperatures(self, T: List[int]) -> List[int]:
        ans = [0] * len(T)
        stack = []
        for i in range(len(T) - 1, -1, -1):
            while stack and T[i] >= T[stack[-1]]:
                stack.pop()
            if stack:
                ans[i] = stack[-1] - i
            stack.append(i)
        return ans
```

```python
#895. Maximum Frequency Stack (Hard)
from collections import Counter

class FreqStack(object):

    def __init__(self):
        self.freq = collections.Counter()
        self.group = collections.defaultdict(list)
        self.maxfreq = 0

    def push(self, x):
        f = self.freq[x] + 1
        self.freq[x] = f
        if f > self.maxfreq:
            self.maxfreq = f
        self.group[f].append(x)

    def pop(self):
        x = self.group[self.maxfreq].pop()
        self.freq[x] -= 1
        if not self.group[self.maxfreq]:
            self.maxfreq -= 1

        return x
```

```python
#996. Number of Squareful Arrays (Hard)
import collections

class Solution:
    def numSquarefulPerms(self, A: List[int]) -> int:
        N = len(A)
        count = collections.Counter(A)

        graph = {x: [] for x in count}
        for x in count:
            for y in count:
                if int((x+y)**.5 + 0.5) ** 2 == x+y:
                    graph[x].append(y)

        def dfs(x, todo):
            count[x] -= 1
            if todo == 0:
                ans = 1
            else:
                ans = 0
                for y in graph[x]:
                    if count[y]:
                        ans += dfs(y, todo - 1)
            count[x] += 1
            return ans

        return sum(dfs(x, len(A) - 1) for x in count)
```
