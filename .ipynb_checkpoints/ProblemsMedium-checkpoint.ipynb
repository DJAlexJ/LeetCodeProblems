```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
    def append(self, element):
        current = self
        if current:
            while current.next:
                current = current.next
            current.next = ListNode(element)
        else:
            self.val = element
            self.next = None
        
    def print_(self):
        current = self
        if current:
            while current.next:
                print(current.val)
                current = current.next
            print(current.val)
        else:
            print('-')
```

```python
#2. Add Two Numbers (Medium)
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        res = []
        for linklist in [l1, l2]:
            res_tmp = []
            current = linklist
            if current:
                while current.next:
                    res_tmp.append(current.val)
                    current = current.next
                res_tmp.append(current.val)
            res.append(res_tmp)
        numbers = []
        for i in range(len(res)):
            n = 0
            for j in range(len(res[i]) - 1, -1, -1):
                n = n * 10 + res[i][j]
            numbers.append(n)
        result = numbers[0] + numbers[1]
        numbers_for_list = [int(d) for d in str(result)]
        numbers_for_list.reverse()
        lst = ListNode(numbers_for_list[0])
        for i in range(1, len(numbers_for_list)):
            lst.append(numbers_for_list[i])
        return lst
```

```python
#8. String To Integer(atoi) (Medium)
import re

int_min = -(2**31)
int_max = 2**31 - 1
        
def form(lst):
    neg = 1
    res = 0
    ind = 0
    if lst[0] == '-':
        neg = -1
        ind += 1
    elif lst[0] == '+':
        neg = 1
        ind += 1
    for i in range(ind, len(lst)):
        if i == 1 and not (lst[i].isdigit()):
            return res
        elif lst[i].isdigit():
            res = res*10 + int(lst[i])
        else:
            res = res * neg
            if res < int_min:
                return int_min
            elif res > int_max:
                return int_max
            return res
    return 0
        
class Solution:
    def myAtoi(self, str: str) -> int:
        arr = re.split(' ', str)
        ind = 0
        while arr[ind] == "" and ind < len(arr):
            ind += 1
            if ind == len(arr):
                return 0
        if re.match('^(-?|\+)\d+(?:\.\d+)?$', arr[ind]) != None:
            res = int(float(arr[ind]))
            if res < int_min:
                return int_min
            elif res > int_max:
                return int_max
            else:
                return int(float(arr[ind]))
        arr = list(arr[ind])
        res = form(arr)
        return res
```

```python
#841. Keys and Rooms (Medium)
class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        seen = [False] * len(rooms)
        seen[0] = True
        stack = [0]
        while stack:
            node = stack.pop()
            for room in rooms[node]:
                if not seen[room]:
                    seen[room] = True
                    stack.append(room)
        return all(seen)
```

```python
#1161. Maximum Level Sum of a Binary Tree (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
def getHeight(root):   
    if (root.left == None and root.right == None): 
        return 0
  
    left = 0
    if (root.left != None): 
        left = getHeight(root.left) 
  
    right = 0
    if (root.right != None): 
        right = getHeight(root.right) 
  
    return (max(left, right) + 1)  
    
class Solution:
    def maxLevelSum(self, root: TreeNode) -> int:
        def calculateLevelSum(root, level):      
            if (root == None): 
                return
  
            sums[level] += root.val 
            calculateLevelSum(root.left, level + 1) 
            calculateLevelSum(root.right, level + 1)
        
        levels = getHeight(root) + 1  
        sums = [0] * levels  
        calculateLevelSum(root, 0)
    
        return max(zip(sums, range(len(sums))))[1] + 1
```

```python
#46. Permutations (Medium)
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        if len(nums) == 0:
            return []
        elif len(nums) == 1:
            return [nums]
        else:
            permut = []
            for i in range(len(nums)):
                m = nums[i]
                remained = nums.copy()
                remained.remove(m)
                for p in self.permute(remained):
                    permut.append([m] + p)
        return permut
```

```python
#94. Binary Tree Inorder Traversal (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        output = []
        stack = []
        while stack or root:
            while root:
                stack.append(root)
                root = root.left
            root = stack.pop()
            output.append(root.val)
            root = root.right
        return output
            
```

```python
#98. Validate Binary Search Tree (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        stack, inorder = [], float('-inf')      
        while stack or root:
            while root:
                stack.append(root)
                root = root.left
            root = stack.pop()
            if root.val <= inorder:
                return False
            inorder = root.val
            root = root.right

        return True
```

```python
#200. Number of Islands (Medium)
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        islands = 0
        for i in range(len(grid)):
            for j in range(len(grid[i])):
                if grid[i][j] == '1':
                    islands += 1
                    self.helper(grid, i, j)
        return islands
    
    def helper(self, grid, i, j):
        if i < 0 or j < 0 or i >= len(grid) or j >= len(grid[i]) or grid[i][j] != '1':
            return
        grid[i][j] = 0
        self.helper(grid, i+1, j)
        self.helper(grid, i-1, j)
        self.helper(grid, i, j+1)
        self.helper(grid, i, j-1)
```

```python
#215. Kth Largest Element in an Array (Medium)
class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        nums.sort(reverse = True)
        return nums[k-1]
```

```python
#654. Maximum Binary Tree (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def constructMaximumBinaryTree(self, nums: List[int]) -> TreeNode:
        if nums == []:
            return None
        max_index = max(zip(nums, range(len(nums))))[1]
        max_elem = nums[max_index]
        root = self.constructHelper(max_elem, nums[:max_index], nums[max_index+1:])
        return root
    
    def constructHelper(self, head, left_nums, right_nums):
        root = TreeNode(head)
        print(root.val)
        if len(left_nums) > 0:
            max_index_l = max(zip(left_nums, range(len(left_nums))))[1]
            head_l = left_nums[max_index_l]
            root.left = self.constructHelper(head_l, left_nums[:max_index_l], left_nums[max_index_l + 1:])
        if len(right_nums) > 0:
            max_index_r = max(zip(right_nums, range(len(right_nums))))[1]
            head_r = right_nums[max_index_r]
            root.right = self.constructHelper(head_r, right_nums[:max_index_r], right_nums[max_index_r + 1:])
        return root
```

```python
#701. Insert into a Binary Search Tree (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def getTargetCopy(original: TreeNode, cloned: TreeNode, target: TreeNode):
    class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        

class Solution:
    def insert_Helper(self, root: TreeNode, val: int):
        if root is None: 
            root = TreeNode(val) 
        else: 
            if root.val < val: 
                if root.right is None: 
                    root.right = TreeNode(val) 
                else: 
                    self.insert_Helper(root.right, val) 
            else: 
                if root.left is None: 
                    root.left = TreeNode(val) 
                else: 
                    self.insert_Helper(root.left, val) 
                    
    def insertIntoBST(self, root: TreeNode, val: int) -> TreeNode:
        self.insert_Helper(root, val)
        return root
    return target
```

```python
#1302. Deepest Leaves Sum (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
        
def height(root, h):
    if root is None:
        return h-1
    else:
        return max(height(root.right, h+1), height(root.left, h+1))

class Solution:
    def deepestLeavesSum(self, root: TreeNode) -> int:
        h = height(root, 1)
    
        def LeavesSumHelper(root, a, cur_h):
            if root == None:
                return 0
            if root.left:
                LeavesSumHelper(root.left, a, cur_h + 1)
            if root.right:
                LeavesSumHelper(root.right, a, cur_h + 1)
            if cur_h == h:
                a.append(root.val)
            
        res = []
        LeavesSumHelper(root, res, 1)
        return sum(res)
```

```python
#1379. Find a Corresponding Node of a Binary Tree in a Clone of That Tree (Medium)
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:
    def getTargetCopy(self, original: TreeNode, cloned: TreeNode, target: TreeNode) -> TreeNode:
        stack = [cloned]
        while(stack):
            cur_node = stack.pop(0)
            
            if cur_node.val == target.val:
                return cur_node            
            else:
                if cur_node.left != None:
                    stack.append(cur_node.left)
                
                if cur_node.right != None:
                    stack.append(cur_node.right)
```
