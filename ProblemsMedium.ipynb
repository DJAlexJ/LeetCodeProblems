```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
        
    def append(self, element):
        current = self
        if current:
            while current.next:
                current = current.next
            current.next = ListNode(element)
        else:
            self.val = element
            self.next = None
        
    def print_(self):
        current = self
        if current:
            while current.next:
                print(current.val)
                current = current.next
            print(current.val)
        else:
            print('-')
```

```python
#2. Add Two Numbers (Medium)
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        res = []
        for linklist in [l1, l2]:
            res_tmp = []
            current = linklist
            if current:
                while current.next:
                    res_tmp.append(current.val)
                    current = current.next
                res_tmp.append(current.val)
            res.append(res_tmp)
        numbers = []
        for i in range(len(res)):
            n = 0
            for j in range(len(res[i]) - 1, -1, -1):
                n = n * 10 + res[i][j]
            numbers.append(n)
        result = numbers[0] + numbers[1]
        numbers_for_list = [int(d) for d in str(result)]
        numbers_for_list.reverse()
        lst = ListNode(numbers_for_list[0])
        for i in range(1, len(numbers_for_list)):
            lst.append(numbers_for_list[i])
        return lst
```

```python
#8. String To Integer(atoi) (Medium)
import re

int_min = -(2**31)
int_max = 2**31 - 1
        
def form(lst):
    neg = 1
    res = 0
    ind = 0
    if lst[0] == '-':
        neg = -1
        ind += 1
    elif lst[0] == '+':
        neg = 1
        ind += 1
    for i in range(ind, len(lst)):
        if i == 1 and not (lst[i].isdigit()):
            return res
        elif lst[i].isdigit():
            res = res*10 + int(lst[i])
        else:
            res = res * neg
            if res < int_min:
                return int_min
            elif res > int_max:
                return int_max
            return res
    return 0
        
class Solution:
    def myAtoi(self, str: str) -> int:
        arr = re.split(' ', str)
        ind = 0
        while arr[ind] == "" and ind < len(arr):
            ind += 1
            if ind == len(arr):
                return 0
        if re.match('^(-?|\+)\d+(?:\.\d+)?$', arr[ind]) != None:
            res = int(float(arr[ind]))
            if res < int_min:
                return int_min
            elif res > int_max:
                return int_max
            else:
                return int(float(arr[ind]))
        arr = list(arr[ind])
        res = form(arr)
        return res
```

```python
#841. Keys and Rooms (Medium)
class Solution:
    def canVisitAllRooms(self, rooms: List[List[int]]) -> bool:
        seen = [False] * len(rooms)
        seen[0] = True
        stack = [0]
        while stack:
            node = stack.pop()
            for room in rooms[node]:
                if not seen[room]:
                    seen[room] = True
                    stack.append(room)
        return all(seen)
```
